### Relevant Background Research


<br> <i><b> Cultural Differences in Game Appreciation: A Study of Player Game Reviews </i> </b>
written by José P. Zagal and Noriko Tomuro, aimed to analyze video game preference in different regions (US/JP) via NLP (natural language processing). This paper collected user reviews and scores from 4 different gaming websites: Metacritic (US critics), Gamespot (US user), Famitsu (JP critics), and Game World (JP users). After further lexical analysis of game review in two different video game communities, there are several distinct different preferences between American and Japanese, especially: game genre selection and user expectation. Talking about games’ genres, words like: “sports”, “action”, and “FPS” will pop up in American games community while Japan gaming websites has more comments related to single player games or RPG (role playing game) with well established story. The famous FPS (first person shooter) games in US: Call of Duty and Mass Effect are not as popular in Japan region, even lots of japanese gamers never heard about it. For user expectations, japanese gamers are harsher and also pay more attention to detail issues. Their comments mainly focus on characteristics motion, bugs, and online lagging problems, even the typo or translation localization problem also have relatively high frequency in websites. Compared to japanese gamers, lexical analysis indicates that american players tend to emphasize replay value of games and has relatively higher bug tolerance. Two groups also have several similar words shows up in frequency table of comments, such as: graphics, music, and control, but the difference in expectation is still unambiguous between those two groups. </br>
